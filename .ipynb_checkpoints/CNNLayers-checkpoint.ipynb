## Import dependencies

```python
import itertools
import matplotlib.pyplot as plt

import os
import cv2
import numpy as np

import tensorflow as tf
from tensorflow.keras.models import load_model, Sequential
```

## Load in CNN model

```python
model=load_model(os.path.join("models","imageclassifier.h5"))
```

```python
model.summary()
```

```python
model.layers
```

```python
# two ways to get a specific layer
# - by regular slicing
# - by calling the "name" from the summary table
print(model.layers[0])
print(model.get_layer("conv2d"))
```

## Extracting layers

```python
intermediate_layers=model.layers[:3]
```

```python
intermediate_model=Sequential(intermediate_layers)
```

```python
intermediate_model.summary()
```

```python
intermediate_model.input_shape
```

## Visualize image

```python
# can use both plt or cv2
# plt
jpgtoarray=plt.imread("happytest.jpg")
print(jpgtoarray.shape)
plt.imshow(jpgtoarray)
```

```python
# cv2
image=cv2.imread("sadtest.jpg")
print(image.shape)
resize=cv2.resize(image, (256,256))
print(resize.shape)
plt.imshow(cv2.cvtColor(resize,cv2.COLOR_BGR2RGB)) #BGR for opencv, RGB for matplotlib
```

## Predict

```python
np.expand_dims(resize,axis=0).shape
```

```python
yhat=intermediate_model.predict(np.expand_dims(resize,axis=0))
```

```python
# output from our intermediate convolutional neural network 
yhat
```

```python
yhat.shape
```

## Visualize the intermediate output

```python
print(yhat.shape)
print(yhat[0].shape)
print(yhat[0,:,:,0].shape)
plt.imshow(yhat[0,:,:,0]) #output from one convolutional neural network layer
```

```python
plt.imshow(yhat[0,:,:,1])
```

```python
num_rows=4
num_cols=4

fig, ax=plt.subplots(ncols=num_cols, nrows=num_rows,figsize=(8,8))
```

```python
# use itertools to get all the possible combinations for the different positions of the subplots
for idx, ax_idx in enumerate(list(itertools.product(range(num_cols),range(num_rows)))):
    print(idx, ax_idx)
```

```python
fig, ax=plt.subplots(ncols=num_cols, nrows=num_rows,figsize=(20,20))
for idx, ax_idx in enumerate(list(itertools.product(range(num_cols),range(num_rows)))):
    ax[ax_idx[0],ax_idx[1]].imshow(yhat[0,:,:,idx])
```

## Trying out different intermediate layers

```python
intermediate_layers=model.layers[:5]
intermediate_layers
```

```python
image=cv2.imread("happytest.jpg")
resize=cv2.resize(image,(256,256))
print(resize.shape)
resize=np.expand_dims(resize,0)
yhat=intermediate_model.predict(resize)
print(yhat.shape)
```
